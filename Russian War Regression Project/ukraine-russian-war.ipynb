```python
personnel = pd.read_csv("../input/2022-ukraine-russian-war/russia_losses_personnel.csv")
personnel.head()
```

```python
personnel.drop(columns={"personnel*","date"},inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
personnel[["POW","personnel"]].plot()
plt.ylabel("Personnel")
plt.xlabel("Day")
plt.title("Russian Personnel Losses")
```

if this war is continous a week more, I hope it doesn't happen. We predict a week with linear regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(personnel[["day"]],personnel["personnel"])
```

```python
prediction = lr.predict(personnel[["day"]])
prediction = pd.DataFrame(prediction)
prediction=prediction.astype(int)
prediction
```

Our model is very good for precition

```python
from sklearn.metrics import r2_score

r2_score(personnel["personnel"],prediction)
```

```python
series = np.arange(100).reshape(-1,1)

prediction = lr.predict(series)
prediction = pd.DataFrame(prediction)
prediction=prediction.astype(int)

prediction
```
