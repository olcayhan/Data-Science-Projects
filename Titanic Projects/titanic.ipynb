```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

```python
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])



model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)
predictions_train = model.predict(X)
```

Evaluation : 

```python
from sklearn.metrics import f1_score
from sklearn.metrics import jaccard_score
print("f1_score : ",f1_score(y, predictions_train, average='weighted') )

print("Jaccard Score : ",jaccard_score(y, predictions_train))
```

Maybe our n_estimators count is not enough or more, we define a function for that.

```python
def n_estimator_count(count):
    model = RandomForestClassifier(n_estimators=count, max_depth=3, random_state=2)
    model.fit(X, y)
    predictions = model.predict(X_test)
    predictions_train1 = model.predict(X)
    
    print("f1_score : %.4f"%f1_score(y, predictions_train1, average='weighted') )
    
```

```python
count = [100,200,500,750,1000]

for i in count:
    n_estimator_count(i)
```

```python
model = RandomForestClassifier(n_estimators=150, max_depth=3, random_state=2)
model.fit(X, y)
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
