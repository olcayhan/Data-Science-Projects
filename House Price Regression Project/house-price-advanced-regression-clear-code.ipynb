```python
ptrain = pd.read_csv("../input/house-prices-advanced-regression-techniques/train.csv")
ptest = pd.read_csv("../input/house-prices-advanced-regression-techniques/test.csv")
```

```python
ptrain.head()
```

```python
ptrain.columns
```

```python
train = ptrain[["MSSubClass","LotFrontage",'LotArea',"Street","OverallQual","OverallCond","SaleCondition","SaleType","YrSold","PoolArea","SalePrice"]]
```

```python
train.shape
```

```python
train.isnull().sum()
```

```python
lotfrange_mean = train["LotFrontage"].mean()
train["LotFrontage"].fillna(lotfrange_mean,inplace=True)
```

```python
lotfrange_mean1 = ptest["LotFrontage"].mean()
ptest["LotFrontage"].fillna(lotfrange_mean1,inplace=True)
```

```python
train.isnull().sum()
```

```python
test = ptest[["MSSubClass","LotFrontage",'LotArea',"Street","OverallQual","OverallCond","SaleCondition","SaleType","YrSold","PoolArea"]]
test
```

```python
street = pd.get_dummies(train["Street"])
train = pd.concat([train,street], axis = 1)
SaleType = pd.get_dummies(train["SaleType"])
train =pd.concat([train,SaleType], axis = 1)
SaleCondition = pd.get_dummies(train["SaleCondition"])
train =pd.concat([train,SaleCondition], axis = 1)
```

```python
train.head()
```

```python
train.drop(["Street","SaleType","SaleCondition"],axis=1,inplace=True)
```

```python
y_train = train["SalePrice"]
train.drop(["SalePrice"],axis=1,inplace=True)
x_train=train
```

```python
x_train
```

```python
street1 = pd.get_dummies(test["Street"])
test = pd.concat([test,street1], axis = 1)
SaleType1 = pd.get_dummies(test["SaleType"])
test =pd.concat([test,SaleType1], axis = 1)
SaleCondition1 = pd.get_dummies(test["SaleCondition"])
test =pd.concat([test,SaleCondition1], axis = 1)
```

```python
test.drop(["Street","SaleType","SaleCondition"],axis=1,inplace=True)
```

```python
test.isnull().sum()
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
x_train.dtypes
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()

regr.fit(x_train, y_train)
```

```python
test_pred = regr.predict(test)
test_pred
```

```python
output = pd.DataFrame({'Id': ptest.Id, 'SalePrice': test_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
